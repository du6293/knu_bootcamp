```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(agefile_path, encoding='cp949', index_col=0)
df=  df.div(df['2021년12월_거주자_총인구수'], axis=0)
del df['2021년12월_거주자_총인구수'], df['2021년12월_거주자_연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요: ')
a = df.index.str.contains(name)
df2 = df[a]
df3 = df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[1:3]]
fig  = plt.figure(figsize=(12,4))
df4 = df3.T.plot()
```
